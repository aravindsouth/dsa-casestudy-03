## Clustering Methods

- KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=121).fit(data)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
data1 = data.copy()
```

```python
data1['kmeans_labels'] = kmeans.labels_
```

```python
data1
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=data1, x="Magnesium", y="OD280", hue="kmeans_labels", palette="deep")
```
